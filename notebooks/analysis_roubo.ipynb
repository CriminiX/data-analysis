### Kmeans

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
df_veic = df_roubo_uteis[[
    'DESCR_MARCA_VEICULO',   
    'DESCR_COR_VEICULO',
    'DESCR_TIPO_VEICULO',
    'BAIRRO',
    'LONGITUDE',
    'LATITUDE'
]].query("""LONGITUDE.notnull() and DESCR_MARCA_VEICULO.notnull() and DESCR_TIPO_VEICULO.notnull() and DESCR_COR_VEICULO.notnull()""")
```

```python
df_veic.isna().sum()
```

```python
df_veic[['LONGITUDE', 'LATITUDE']] = df_veic[['LONGITUDE', 'LATITUDE']].replace(r',', '.', regex=True)\
  .astype({'LONGITUDE': 'double', 'LATITUDE': 'double'})
```

```python
categorical = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(drop='first')
)

numerical = make_pipeline(
    StandardScaler()
)
```

```python
categorical_cols = ['DESCR_MARCA_VEICULO', 'DESCR_COR_VEICULO', 'DESCR_TIPO_VEICULO', 'BAIRRO']
numerical_cols   = ['LATITUDE', 'LONGITUDE']

tranformer = ColumnTransformer(transformers=[
    ('categorical', categorical, categorical_cols),
    ('numerical', numerical, numerical_cols)
])
```

```python
X = tranformer.fit_transform(df_veic)
```

```python
# Pegando apenas longitude e latitude
X = df_veic.iloc[:, [4, 5]]

n_clusters = 10
kmeans = KMeans(
    n_clusters=n_clusters,
    random_state=0
)

kmeans.fit(X)
labels = kmeans.labels_
X['cluster'] = labels
```

```python
labels
```

```python
_clusters = X.groupby('cluster').count()
# _clusters
_clusters
```

```python
X = X.reset_index()

ptsymb = np.array(['b.','r.','m.','g.','c.','k.','b*','r*','m*','r^']);
plt.figure(figsize=(5,5))
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
for i in range(n_clusters):
    cluster=np.where(labels==i)[0]
    plt.plot(X['LONGITUDE'][cluster].values, X['LATITUDE'][cluster].values, ptsymb[i])
plt.show()
```

```python
X = X.reset_index(drop=True)

ptsymb = np.array(['b.','r.','m.','g.','c.','k.','b*','r*','m*','r^']);
plt.figure(figsize=(5,5))
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
for i in range(n_clusters):
    cluster=np.where(labels==i)[0]
    plt.plot(X['LONGITUDE'][cluster].values, X['LATITUDE'][cluster].values)
plt.show()
```
